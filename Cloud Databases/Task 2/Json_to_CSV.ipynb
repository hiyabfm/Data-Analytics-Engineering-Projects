```python
import json
import csv

# Load JSON data from file
with open('TransactionData.json', 'r') as file:
    data = json.load(file)

# Prepare sets and lists to collect data
customers = {}
transactions = []
transaction_items = []
products = {}

# Process each transaction
for txn in data:
    customer = txn["customer"]
    customer_id = customer["customer_id"]
    
    # Add customer if not already added
    if customer_id not in customers:
        customers[customer_id] = {
            "customer_id": customer_id,
            "name": customer["name"],
            "email": customer["email"],
            "phone": customer["phone"],
            "address": customer["address"]
        }

    # Add transaction
    transactions.append({
        "transaction_id": txn["transaction_id"],
        "customer_id": customer_id,
        "transaction_date": txn["transaction_date"],
        "shipping_method": txn["shipping_method"],
        "total_amount": txn["total_amount"]
    })

    # Add transaction items and products
    for item in txn["shopping_cart"]:
        # Clean the key "vendor: " which has a space
        vendor = item.get("vendor: ", item.get("vendor", "")).strip()
        
        product_id = item["product_id"]

        # Add product if not already added
        if product_id not in products:
            products[product_id] = {
                "product_id": product_id,
                "vendor": vendor,
                "product_category": item["product_category"],
                "product_name": item["product_name"]
            }

        # Add transaction item
        transaction_items.append({
            "transaction_id": txn["transaction_id"],
            "product_id": product_id,
            "quantity": item["quantity"],
            "price": item["price"]
        })

# Write to CSV files
def write_csv(filename, fieldnames, rows):
    with open(filename, 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)

# CUSTOMERS.csv
write_csv('CUSTOMERS.csv', ["customer_id", "name", "email", "phone", "address"], customers.values())

# TRANSACTIONS.csv
write_csv('TRANSACTIONS.csv', ["transaction_id", "customer_id", "transaction_date", "shipping_method", "total_amount"], transactions)

# TRANSACTION_ITEMS.csv
write_csv('TRANSACTION_ITEMS.csv', ["transaction_id", "product_id", "quantity", "price"], transaction_items)

# PRODUCTS.csv
write_csv('PRODUCTS.csv', ["product_id", "vendor", "product_category", "product_name"], products.values())
```
