```python
import pandas as pd

# Load the CSV with all columns as strings
df = pd.read_csv('CUSTOMERS.csv', dtype=str)

# Strip whitespace and standardize each column
df['customer_id'] = df['customer_id'].str.strip()

df['name'] = df['name'].str.strip().str.title()  # e.g., "russell anderson" → "Russell Anderson"

df['email'] = df['email'].str.strip().str.lower()  # normalize email to lowercase

df['phone'] = df['phone'].str.strip().str.replace(r'\D', '', regex=True)  # keep only digits

# Standardize address:
# - Strip whitespace
# - Replace multiple spaces or line breaks with a single space
df['address'] = (
    df['address']
    .str.strip()
    .str.replace(r'\s+', ' ', regex=True)  # normalize spaces/newlines
)

# Optional: show result
print(df.head())
```

```python
df.to_csv('CUSTOMERS.csv', index=False)
```
