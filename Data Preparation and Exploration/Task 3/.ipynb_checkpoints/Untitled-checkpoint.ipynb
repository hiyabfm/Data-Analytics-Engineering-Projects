```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv('Megastore_Dataset_Task_3 3.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
# Data Cleaning
df['ProductName'] = df['ProductName'].str.strip() #this is to remove spaces
```

```python
df['OrderID'] = df['OrderID'].astype('str') #convert Orrder ID to string since i wont be using it for mathimatical cal
```

```python
df['CustomerOrderSatisfaction'].unique()
```

```python
df['OrderPriority'].unique()
```

```python
# Ordinal Encoding 
priority_map = {'Low': 0, 'Medium': 1, 'High': 2}
satisfaction_map = {'Prefer not to answer': 0,'Very Dissatisfied': 1, 'Dissatisfied': 2, 'Satisfied': 3, 'Very Satisfied': 4}
df['OrderPriority'] = df['OrderPriority'].map(priority_map)
df['CustomerOrderSatisfaction'] = df['CustomerOrderSatisfaction'].map(satisfaction_map)
```

```python
# One Hot encoding
df = pd.get_dummies(df, columns=['Region', 'Segment'])
```

```python
df.head()
```

```python
#Transactionalizing

basket = (
    df.groupby(['OrderID', 'ProductName'])['Quantity']
      .sum().unstack().reset_index().fillna(0)
      .set_index('OrderID')
)
```

```python
basket.head()
```

```python
def encode_data(x):
    return 1 if x > 0 else 0

basket_data = basket.stack().map(lambda x: 1 if x > 0 else 0).unstack()
basket_data = basket_data.astype(bool)
```

```python
display (basket_data)
```

```python
df.to_csv('Encoded_Megastore_Dataset_Task_3.csv',index=False)
```

```python
basket_data.to_csv('Cleaned_Megastore_Dataset_Task_3.csv',index=False)
```

```python
# Run Aprioro Algoritm
frequent_itemsets = apriori(basket_data, min_support=0.09, use_colnames=True)
# data is structured and i can start generateing frequest item sets that have a support of at least 10%
```

```python
# Create Rule
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```

```python
# Filter and sort key metrics
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules = rules.sort_values(by='lift', ascending=False)
display (rules)
```
