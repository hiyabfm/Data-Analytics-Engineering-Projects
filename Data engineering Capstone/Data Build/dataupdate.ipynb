```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("washington_daily_weather_2024.csv")

# Dictionary of city populations
population_map = {
    "Seattle": 816600,
    "Spokane": 230609,
    "Tacoma": 228202,
    "Vancouver": 198992,
    "Bellevue": 154377,
    "Everett": 113000,
    "Olympia": 55605,
    "Bellingham": 96701,
    "Yakima": 97878,
    "Kennewick": 87549
}

weekend_map = {
    "Yes": 2,
    "No": 1
}

season_map = {
    "Winter": 2,
    "Summer": 5,
    "Autumn": 4,
    "Spring": 3
}

df["weekend_map"]= df["weekend"].map(weekend_map)
df["season_map"]= df["season"].map(season_map)
df["prcp"] = df["prcp"].fillna(0.0)

# Create the new column based on city matches
df["estimated_population"] = df["city"].map(population_map)
df["daily_sales"] = ((df["estimated_population"] / 1000) * df["weekend_map"] * df["season_map"]) + df["tavg"] + df["prcp"]
df["daily_sales"] = df["daily_sales"].round(2)

# Save the updated CSV
df.to_csv("washington_daily_weather_2024_with_population.csv", index=False)

print("✅ New file created: washington_daily_weather_2024_with_population.csv")
```
