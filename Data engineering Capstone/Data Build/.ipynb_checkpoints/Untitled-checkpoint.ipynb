```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("washington_daily_weather_2024_with_population.csv")

# ✅ Drop unnecessary columns if they exist
columns_to_drop = ["wdir", "wpgt", "tsun", "weekend_map", "season_map", "weekend", "snow"]
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

# Save the cleaned dataset
df.to_csv("washington_daily_auto_sales.csv", index=False)

print("✅ Columns removed and new file created: washington_daily_auto_sales.csv")
```

```python
import pandas as pd

# Load your existing CSV file
csv_file = "washington_x_restaurant_daily_sales_dataset.csv"
df = pd.read_csv(csv_file)

# Save it as an Excel file
excel_file = "washington_x_restaurant_daily_sales_dataset.xlsx"
df.to_excel(excel_file, index=False)

print(f"✅ Successfully converted {csv_file} → {excel_file}")
```
