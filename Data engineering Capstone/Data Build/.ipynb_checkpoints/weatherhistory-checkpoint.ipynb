```python
# Install the meteostat package if you haven't already
# !pip install meteostat

from datetime import datetime
from meteostat import Point, Daily

# Define the time period (example: full year 2024)
start = datetime(2024, 1, 1)
end   = datetime(2024, 12, 31)

# Define a location in Alaska — let's use Anchorage (latitude, longitude, elevation)
alaska = Point(61.2181, -149.9003, 31)  # Anchorage, Alaska

# Get daily weather data
data = Daily(alaska, start, end)
df = data.fetch()

# Display first few rows
print(df.head())

# Optional: show column names
print("\nColumns:", df.columns.tolist())

```

```python
# Save to CSV file
output_path = "alaska_daily_weather_2024.csv"
df.to_csv(output_path, index=True)  # index=True keeps the date column

print(f"\n✅ Data saved successfully to: {output_path}")
```

```python
# !pip install meteostat pandas

from datetime import datetime
from meteostat import Point, Daily
import pandas as pd

# Define time range
start = datetime(2023, 1, 1)
end   = datetime(2025, 6, 1)

# Define major Washington cities with coordinates (lat, lon, elevation in meters)
cities = {
    "Seattle": Point(47.6062, -122.3321, 56),
    "Spokane": Point(47.6588, -117.4260, 562),
    "Tacoma": Point(47.2529, -122.4443, 74),
    "Vancouver": Point(45.6387, -122.6615, 52),
    "Bellevue": Point(47.6104, -122.2007, 85),
    "Everett": Point(47.9780, -122.2021, 25),
    "Olympia": Point(47.0379, -122.9007, 58),
    "Bellingham": Point(48.7519, -122.4787, 30),
    "Yakima": Point(46.6021, -120.5059, 325),
    "Kennewick": Point(46.2112, -119.1372, 120)
}

# Create an empty list to hold all data
all_data = []

# Loop through cities and fetch daily weather data
for city, location in cities.items():
    print(f"Fetching weather data for {city}...")
    data = Daily(location, start, end).fetch()
    data['city'] = city  # Add city name column
    all_data.append(data)

# Combine all city data into one DataFrame
df_all = pd.concat(all_data)

# Reset index to keep date as a column
df_all.reset_index(inplace=True)

# 🟢 ADD MONTH, SEASON, AND WEEKEND COLUMNS
df_all['month'] = df_all['time'].dt.month
df_all['month_name'] = df_all['time'].dt.strftime('%B')

# Define a helper function to assign seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df_all['season'] = df_all['month'].apply(get_season)

# 🆕 Add weekend column (Yes if Saturday or Sunday)
df_all['weekend'] = df_all['time'].dt.dayofweek.apply(lambda x: 'Yes' if x >= 5 else 'No')

# Save to CSV
output_file = "washington_daily_weather_2024.csv"
df_all.to_csv(output_file, index=False)

print(f"\n✅ Daily weather data for major Washington cities saved to: {output_file}")
print("🗓️ Added columns: month, month_name, season, weekend")
```

```python
# time → date of observation

# tavg → average daily temperature (°C)

# tmin → minimum daily temperature (°C)

# tmax → maximum daily temperature (°C)

# prcp → total precipitation (mm)

# snow → snowfall (mm)

# wspd → average wind speed (km/h or m/s depending on Meteostat units)

# pres → average atmospheric pressure (hPa)

# tsun → sunshine duration (minutes)

# city → name of the city (e.g., “Seattle”)

# month → numeric month (1–12)

# month_name → full month name (“January”, “February”, etc.)

# season → season (“Winter”, “Spring”, “Summer”, “Autumn”)
```
