```python
import pandas as pd

# Load both datasets
weather_df = pd.read_csv("washington_daily_weather_2024_with_population.csv")
sales_df = pd.read_csv("x_restaurant_sales.csv")

# Inspect columns to confirm the shared key
print(weather_df.columns)
print(sales_df.columns)

# Merge (join) the datasets on the 'time' column
# Use an inner join to keep only matching rows, or change to 'left'/'right'/'outer' as needed
merged_df = pd.merge(sales_df, weather_df, on='time', how='inner')

# Display the first few rows of the merged dataset
print(merged_df.head())

# (Optional) Save to a new CSV file
merged_df.to_csv("washington_x_restaurant_daily_sales_datasets.csv", index=False)
```
