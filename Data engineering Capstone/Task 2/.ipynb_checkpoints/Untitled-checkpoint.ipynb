```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge, LassoCV, BayesianRidge
import seaborn as sns
import matplotlib.pylab as plt
import numpy as np

from dmba import regressionSummary, exhaustive_search
from dmba import backward_elimination, forward_selection, stepwise_selection
from dmba import adjusted_r2_score, AIC_score, BIC_score

```

```python
df = pd.read_csv('washington_x_restaurant_daily_sales_dataset.csv')
```

```python
df.head()
```

```python
df['season'] = df['season'].map({'Winter': 1, 'Spring': 2, 'Summer': 3, 'Autumn':4})
```

```python
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'])

# Extract year
df['year'] = df['time'].dt.year

# Extract day of year (1 to 365/366)
df['day_of_year'] = df['time'].dt.dayofyear

df = df.drop(columns=['time', 'month_name'])
```

```python
df.head()
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

variables = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'month', 'season', 'estimated_population', 'year', 'day_of_year', 'daily_sales','city']

for feature in variables:
    plt.figure(figsize=(10, 2))
    sns.scatterplot(x=df[feature], y=df['daily_sales'])
    plt.title(f'daily_sales vs. {feature}')
    plt.xlabel(feature)
    plt.ylabel('daily_sales')
    plt.tight_layout()
    # plt.savefig(f'scatter_{feature}.png')  # Saves figure as PNG
    plt.show()
```

```python
df[variables].head()
```

```python
df_variables = df[variables]

# One-hot encode 'city' (drop_first=True avoids dummy variable trap)
df_encoded = pd.get_dummies(df_variables, columns=['city'], drop_first=True)

print(df_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()
features_std = scaler.fit_transform(df_encoded)  # df_encoded is already the subset of features
features_std_df = pd.DataFrame(features_std, columns=df_encoded.columns, index=df_encoded.index)

print(features_std_df.head())
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=12)
features_pca = pca.fit_transform(features_std_df)
```

```python
print(pca.explained_variance_ratio_)
```

```python
explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(8, 4))
plt.bar(range(1, len(explained_variance)+1), explained_variance, alpha=0.7)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components')
plt.grid(True)
plt.show()
```

```python
df_variables = df[variables]  # variables is a list of column names including 'daily_sales'

# One-hot encode 'city' only, leave other numeric columns untouched
final_variables = pd.get_dummies(df_variables, columns=['city'], drop_first=True).astype(int)

print(final_variables.head())
```

```python
# Separate features (X) and target (y)
x = final_variables.drop(columns=['daily_sales'])  # all columns except 'daily_sales'
y = final_variables['daily_sales']                # target variable

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lreg = LinearRegression()
lreg.fit(X_train, y_train)
```

```python
print(pd.DataFrame({'features': x.columns, 'coefficient': lreg.coef_}))
regressionSummary(y_train, lreg.predict(X_train))
```

```python
def train_model(variables):
    model = LinearRegression()
    model.fit(X_train[variables], y_train)
    return model

def score_model(model, variables):
    return AIC_score(y_train, model.predict(X_train[variables]), model)
    
allVariables = list(X_train.columns)

best_model, best_variables = backward_elimination(allVariables, train_model, score_model, verbose=True)

print(best_variables)

regressionSummary(y_test, best_model.predict(X_test[best_variables]))
```

```python
best_model, best_variables = stepwise_selection(
    list(X_train.columns),
    train_model,
    score_model,
    direction='backward', 
    verbose=True
)
print(best_variables)
```

```python
import statsmodels.api as sm
X_train_const = sm.add_constant(X_train[best_variables])
ols_model = sm.OLS(y_train, X_train_const).fit()

print(ols_model.summary())
```

```python
import statsmodels.formula.api as smf

# Combine training features and target into one DataFrame
train_df = X_train.join(y_train)

# Use the correct target column name
target = 'daily_sales'  # change this to your actual target column

# Build the formula string
features = X_train.columns
formula = target + ' ~ ' + ' + '.join(features)

# Fit the OLS model
feature_lm = smf.ols(formula=formula, data=train_df).fit()

# View summary
feature_lm.summary()
```

```python
from sklearn.metrics import mean_squared_error

y_train_pred = ols_model.predict(X_train_const)
mse_train = mean_squared_error(y_train, y_train_pred)

print(f"Training MSE: {mse_train}")
```

```python
X_test_const = sm.add_constant(X_test[best_variables])
y_test_pred = ols_model.predict(X_test_const)

mse_test = mean_squared_error(y_test, y_test_pred)
print(f"Test MSE: {mse_test}")
```
