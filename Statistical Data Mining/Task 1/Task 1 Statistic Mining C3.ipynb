```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge, LassoCV, BayesianRidge
import seaborn as sns
import matplotlib.pylab as plt
import numpy as np

from dmba import regressionSummary, exhaustive_search
from dmba import backward_elimination, forward_selection, stepwise_selection
from dmba import adjusted_r2_score, AIC_score, BIC_score

```

```python
df = pd.read_csv('D600 Task 1 Dataset 1 Housing Information.csv')
```

```python
df['Garage'] = df['Garage'].map({'Yes': 1, 'No': 0})
```

```python
features = ['Price', 'SquareFootage', 'NumBathrooms', 'CrimeRate', 'SchoolRating',
            'RenovationQuality', 'LocalAmenities', 'DistanceToCityCenter', 'Garage']
df[features].describe()
```

```python
df[features].head()
```

```python
%matplotlib inline
sns.set(style="whitegrid")

for feature1 in features:
    plt.figure(figsize=(4, 2))
    sns.histplot(df[feature1], kde=True)
    plt.title(f'Distribution of {feature1}')
    plt.xlabel(feature1)
    plt.ylabel('Frequency')
    plt.show()
```

```python
%matplotlib inline
for feature2 in features[1:]:  # Exclude 'Price'
    plt.figure(figsize=(4, 2))
    sns.scatterplot(x=df[feature2], y=df['Price'])
    plt.title(f'Price vs. {feature2}')
    plt.xlabel(feature2)
    plt.ylabel('Price')
    plt.show()
```
